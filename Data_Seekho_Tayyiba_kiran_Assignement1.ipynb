# **Exercise 6: Built-in Functions**⏪⏩⏬⏫

```python
# Taking a list of numbers as input
user_input = input("Enter a list of numbers separated by commas: ")
number_list = [float(num) for num in user_input.split(',')]
 # Convert input to a list of floats

# Using built-in functions
absolute_values = [abs(num) for num in number_list]
maximum_value = max(number_list)
minimum_value = min(number_list)

# Printing results
print("Absolute values:", absolute_values)
print("Maximum value:", maximum_value)
print("Minimum value:", minimum_value)
```

# **Exercise 7: Arithmetic Operations**➕➖✖➗

```python
# Simple calculator program
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Selecting an operation
operation = input("Select an operation (add, subtract, multiply, divide): ").strip().lower()

# Performing the selected operation
if operation == "add":
    result = num1 + num2
elif operation == "subtract":
    result = num1 - num2
elif operation == "multiply":
    result = num1 * num2
elif operation == "divide":
    try:
        result = num1 / num2
    except ZeroDivisionError:
        result = "Error: Division by zero is not allowed."
else:
    result = "Invalid operation selected."

# Printing the result
print("Result:", result)
```
