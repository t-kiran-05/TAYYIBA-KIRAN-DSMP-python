# **ðŸš€Overview of the Course Management System CodeðŸ›ï¸**

This code implements a course management system using object-oriented programming (OOP) principles in Python. It consists of a base class Course and two derived classes, ProgrammingCourse and MathematicsCourse. Below is a concise overview highlighting the OOP concepts utilized in the code.

# **1. Class StructureðŸ›ï¸**
## **Course Class:**

**Attributes:**
__course_name, __credits, and __instructor: Private attributes encapsulating course details.

**Methods:**

get_course_name(): Returns the course name.

get_credits(): Returns the number of credits.

get_instructor(): Returns the instructor's name.

get_syllabus(): Returns a syllabus statement.

enroll_student(student_name): Enrolls a student in the course and returns a confirmation message.

## **ProgrammingCourse Class:**

Inherits from Course.

programming_language: Specifies the programming language for the course.

**Methods:**

get_practical_exercises(): Returns details about practical exercises.

get_schedule(): Overrides the method to provide a specific schedule for programming courses.

**MathematicsCourse Class:**

Inherits from Course.
Additional Attribute:
math_field: Specifies the mathematical field of study.

**Methods:**

get_formula_sheet(): Returns a formula sheet for the course.

get_schedule(): Overrides the method to provide a specific schedule for mathematics courses.

# **2. OOP Concepts Used**
# Encapsulation:

Private attributes in the Course class protect the internal state of the object from direct access, promoting data hiding.

# Inheritance:

ProgrammingCourse and MathematicsCourse inherit from the Course class, allowing them to reuse and extend the functionality of the base class.

# Polymorphism:

The get_schedule() method is overridden in both derived classes to provide specific implementations for programming and mathematics courses, demonstrating polymorphic behavior.

# **3. User Interaction**
The main method allows users to input details for both programming and mathematics courses interactively.

After creating course objects, it displays relevant **course information, including enrollment confirmation and specific course-related details.**










```python
class Course:
    def __init__(self, course_name, credits, instructor):
        self.__course_name = course_name  # Private attribute
        self.__credits = credits            # Private attribute
        self.__instructor = instructor      # Private attribute

    # Public methods to access private attributes
    def get_course_name(self):
        return self.__course_name

    def get_credits(self):
        return self.__credits

    def get_instructor(self):
        return self.__instructor

    def get_syllabus(self):
        return f"Syllabus for {self.__course_name}"

    def enroll_student(self, student_name):
        return f"{student_name} has been enrolled in {self.__course_name}"


class ProgrammingCourse(Course):
    def __init__(self, course_name, credits, instructor, programming_language):
        super().__init__(course_name, credits, instructor)
        self.programming_language = programming_language  # Additional attribute

    def get_practical_exercises(self):
        return f"Practical exercises for {self.get_course_name()} in {self.programming_language}"

    def get_schedule(self):
        return f"{self.get_course_name()} includes coding assignments and lab sessions."


class MathematicsCourse(Course):
    def __init__(self, course_name, credits, instructor, math_field):
        super().__init__(course_name, credits, instructor)
        self.math_field = math_field  # Additional attribute

    def get_formula_sheet(self):
        return f"Formula sheet for {self.get_course_name()} in {self.math_field}"

    def get_schedule(self):
        return f"{self.get_course_name()} includes theory lectures and problem-solving sessions."


# Main method to run the project
if __name__ == "__main__":
    # Taking input for a programming course
    print("Enter details for Programming Course:")
    programming_course_name = input("Course Name: ")
    programming_credits = int(input("Credits: "))
    programming_instructor = input("Instructor: ")
    programming_language = input("Programming Language: ")

    programming_course = ProgrammingCourse(programming_course_name, programming_credits, programming_instructor, programming_language)

    # Taking input for a mathematics course
    print("\nEnter details for Mathematics Course:")
    mathematics_course_name = input("Course Name: ")
    mathematics_credits = int(input("Credits: "))
    mathematics_instructor = input("Instructor: ")
    math_field = input("Mathematics Field (e.g., Calculus, Statistics): ")

    mathematics_course = MathematicsCourse(mathematics_course_name, mathematics_credits, mathematics_instructor, math_field)

    # Displaying course information
    print("\nCourse Information:")
    print("-------------------")

    # Display Programming Course Info
    print(f"Course Name: {programming_course.get_course_name()}")
    print(f"Instructor: {programming_course.get_instructor()}")
    print(f"Credits: {programming_course.get_credits()}")
    print(programming_course.get_syllabus())
    student_name = input("Enter student name to enroll in Programming Course: ")
    print(programming_course.enroll_student(student_name))
    print(programming_course.get_practical_exercises())
    print(programming_course.get_schedule())
    print()

    # Display Mathematics Course Info
    print(f"Course Name: {mathematics_course.get_course_name()}")
    print(f"Instructor: {mathematics_course.get_instructor()}")
    print(f"Credits: {mathematics_course.get_credits()}")
    print(mathematics_course.get_formula_sheet())
    student_name = input("Enter student name to enroll in Mathematics Course: ")
    print(mathematics_course.enroll_student(student_name))
    print(mathematics_course.get_schedule())
```
