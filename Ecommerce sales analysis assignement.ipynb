# EDA

Monthly Sales

```python
sales.columns
```

```python
orders.columns
```

```python
# prompt: extract month from OrderDate

# Convert 'OrderDate' to datetime objects
#orders['OrderDate'] = pd.to_datetime(orders['OrderDate'])

# Extract the month from the 'OrderDate' column
orders['Month'] = orders['OrderDate'].dt.month

# Print the updated dataframe to see the new 'Month' column
orders
```

```python
monthly_sales = sales.groupby('Month')['Revenue'].sum()
print(monthly_sales)
```

Merge Sales and Orders Files

```python
merge_sales_orders = pd.merge(sales, orders, left_on='OrderID', right_on='TransactionID')
merge_sales_orders
```

```python
monthly_sales=merge_sales_orders.groupby(orders['Month'])['Revenue'].sum()
print(monthly_sales)
```

Customers Behavior

```python
top_customers = merge_sales_orders.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False)
print(top_customers.head(5))
```

Product Performance

```python
top_products = merge_sales_orders.groupby('ProductID')['Revenue'].sum().sort_values(ascending=False)
print(top_products.head(5))
```
